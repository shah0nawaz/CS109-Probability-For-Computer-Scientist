How many unique 6-digit passcodes are possible if a
phone password uses each of four distinct numbers?
Two mutually exclusive scenarios:
 One digit repeated three times, other three repeated once
 Two digits repeated twice, other two repeated once

```python
# solution 
# case.1

total = 4*math.factorial(6)/(math.factorial(3))
total
```

```python
# solution 
# Case 2

total = 6*math.factorial(6)/(math.factorial(2)* math.factorial(2))
total
```

## Combinations	with	cake
There are n = 20 people.
How many ways can we choose  = 5 people to get cake?

Solution:

We first overcount the possible orderings of 20 people considering the order.
Secondly, as we have to chose k people from those 20 to give them cakes, as in this case the order of the persons with cakes does not matter. 

overcounting n! ways
as we are overcounting the k! ways of those chosen people with cakes and also overcounting those of (n-k)! with no cakes
total = # of total persons! / remove the # number of selected people with cakes ! * remove the # number of unseleted people with no cakes 

total = n!/ (n-k)! k!  bionomial coefficient 





```python
total = math.factorial(20)/math.factorial(5)*(math.factorial(15))
total
```
