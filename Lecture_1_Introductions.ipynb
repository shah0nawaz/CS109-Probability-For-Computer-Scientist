Example 2. How many CA license plates are possible with each of the following formats?

SAM 123

6FGR999

In the first case we have three characters positions and three digits positions. 
For each character position we 26 choices [a-z] and for each decimal digit 
position we 10 choices [0-9]. Using the product rule of counting we multiply them all.

```python
# first case
total = (26**3)*(10**3)
print(f'Total number of possible number plates {total}')
```

In the second case we have three characters positions and four digits positions. 
For each character position we 26 choices [a-z] and for each decimal digit 
position we 10 choices [0-9]. Using the product rule of counting we multiply them all.

```python
# Second case
total = (9+1)*(26**3)*(10**3)
print(f'Total number of possible number plates {total}')
```
